```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import TimeoutException, NoSuchElementException
import time

driver = webdriver.Chrome()

driver.get("https://account.umbler.com/SignIn?App=UTalk2")
driver.maximize_window()

# LOGIN
time.sleep(5)
email_elem = driver.find_element(By.XPATH, "/html/body/div[1]/div[1]/div/form/div[1]/input")
email_elem.send_keys("filipe.hernandes@sorriabank.com.br")

senha_elem = driver.find_element(By.XPATH, "/html/body/div[1]/div[1]/div/form/div[2]/div/input")
senha_elem.send_keys("Sorria2020$")

btn_elem = driver.find_element(By.XPATH, "/html/body/div[1]/div[1]/div/form/input[4]")
btn_elem.click()

time.sleep(6)

try:
    # Espera até que o elemento esteja presente na página e visível
    elemento_de_merda = WebDriverWait(driver, 30).until(
        EC.visibility_of_element_located((By.XPATH, '/html/body/div[47]/div'))
    )

    if elemento_de_merda:
        # Remove o elemento da página
        driver.execute_script("arguments[0].remove();", elemento_de_merda)

except TimeoutException:
    print("O elemento não foi encontrado dentro do tempo esperado.")

time.sleep(3)
criar_elem = driver.find_element(By.XPATH, "/html/body/section/div/div/div/div[2]/div/div[1]/div[1]/span[2]/button")
driver.execute_script("arguments[0].click();", criar_elem)

novo_contato = driver.find_element(By.XPATH, "/html/body/section/div/div/div/div[3]/div/div/div[2]/div[2]/div[2]/button")
driver.execute_script("arguments[0].click();", novo_contato)
try:
    # Verifique se o modal está visível
    WebDriverWait(driver, 30).until(
        EC.visibility_of_element_located((By.CSS_SELECTOR, 'div.modal-content'))
    )

    # Use JavaScript para definir os valores dos campos
    driver.execute_script("""
        setTimeout(function() {
            var nomeContato = document.querySelector('input[id="EE756FBB40D822B0487A"]');
            var celularContato = document.querySelector('input.form-control.iti__tel-input');

            if (nomeContato) {
                nomeContato.value = 'TESTE';
                nomeContato.dispatchEvent(new Event('input', { bubbles: true }));
            } else {
                console.log('Elemento nomeContato não encontrado');
            }

            if (celularContato) {
                celularContato.value = '11999999999';
                celularContato.dispatchEvent(new Event('input', { bubbles: true }));
            } else {
                console.log('Elemento celularContato não encontrado');
            }
        }, 1000);  // 1 segundo de delay
    """)

except TimeoutException:
    print("O elemento não foi encontrado dentro do tempo esperado.")
except NoSuchElementException as e:
    print(f"Elemento não encontrado: {e}")
```
