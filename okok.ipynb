```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import TimeoutException
import time

driver = webdriver.Chrome()

driver.get("https://account.umbler.com/SignIn?App=UTalk2")

driver.maximize_window()

#LOGIN
time.sleep(5)
email_elem = driver.find_element(By.XPATH, "/html/body/div[1]/div[1]/div/form/div[1]/input")
email_elem.send_keys("filipe.hernandes@sorriabank.com.br")

senha_elem = driver.find_element(By.XPATH, "/html/body/div[1]/div[1]/div/form/div[2]/div/input")
senha_elem.send_keys("Sorria2020$")

btn_elem = driver.find_element(By.XPATH, " ")
btn_elem.click()

time.sleep(2)

try:
    # Espera até que o elemento esteja presente na página
    elemento_de_merda = WebDriverWait(driver, 30).until(
        EC.presence_of_element_located((By.XPATH, '/html/body/div[47]/div'))
    )

    if elemento_de_merda:
        # Torna o elemento invisível usando JavaScript
        driver.execute_script("arguments[0].style.visibility='hidden'", elemento_de_merda)

except TimeoutException:
    print("O elemento não foi encontrado dentro do tempo esperado.")


time.sleep(3)
criar_elem = driver.find_element(By.XPATH, "/html/body/section/div/div/div/div[2]/div/div[1]/div[1]/span[2]/button")
driver.execute_script("arguments[0].click();", criar_elem)

time.sleep(3)

novo_contato = driver.find_element(By.XPATH, "/html/body/section/div/div/div/div[3]/div/div/div[2]/div[2]/div[2]/button")
driver.execute_script("arguments[0].click();", novo_contato)

nome_contato = driver.find_element(By.XPATH, "/html/body/section/div[2]/div/div/div[3]/div/div/div[2]/div/div[2]/div/form/div[1]/div[1]/div[1]/input")
nome_contato.send_keys("TESTE")

senha_contato = driver.find_element(By.XPATH, "/html/body/section/div[2]/div/div/div[3]/div/div/div[2]/div/div[2]/div/form/div[1]/div[1]/div[2]/div/div/input")
senha_contato.send_keys('(11)99999-9999')

email_elem = driver.find_element(By.XPATH, "/html/body/section/div[2]/div/div/div[3]/div/div/div[2]/div/div[2]/div/form/div[1]/div[2]/div[2]/input")
email_elem.send_keys('mecaguei@gmail.com')

salvar_contato = driver.find_element(By.XPATH, "/html/body/section/div[2]/div/div/div[3]/div/div/div[2]/div/div[2]/div/form/div[2]/div/button[2]")
driver.execute_script("arguments[0].click();", salvar_contato)
```
