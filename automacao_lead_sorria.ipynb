```python
def get_mouse_position_percentage():
    screen_width, screen_height = pa.size()
    current_mouse_x, current_mouse_y = pa.position()
    x_percentage = (current_mouse_x / screen_width) * 100
    y_percentage = (current_mouse_y / screen_height) * 100
    return x_percentage, y_percentage
get_mouse_position_percentage()

def move_mouse_to_percentage(x_percentage, y_percentage):
    screen_width, screen_height = pa.size()
    x_position = screen_width * (x_percentage / 100)
    y_position = screen_height * (y_percentage / 100)
    pa.moveTo(x_position, y_position)
time.sleep(2)
mouse_x_percentage, mouse_y_percentage = get_mouse_position_percentage()
print(mouse_x_percentage, mouse_y_percentage)
```
